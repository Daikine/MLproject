# EDA: синтетические продажи (Milestone 1)

Ноутбук показывает:
- структуру данных
- графики продаж по SKU
- недельную сезонность
- эффект промо


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/sales.csv', parse_dates=['date'])
df.head()
```

```python
# График продаж по одному SKU
sku = df['sku'].unique()[0]
g = df[df['sku']==sku].sort_values('date')
plt.figure()
plt.plot(g['date'], g['sales'])
plt.title(f'Sales for {sku}')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Недельная сезонность (среднее по дню недели)
df['dow'] = df['date'].dt.dayofweek
weekly = df.groupby('dow')['sales'].mean()
plt.figure()
plt.bar(weekly.index, weekly.values)
plt.title('Weekly seasonality (mean sales by day-of-week)')
plt.xlabel('Day of week (0=Mon)')
plt.ylabel('Mean sales')
plt.show()
```

```python
# Эффект промо
if 'promo_flag' in df.columns:
    promo = df.groupby('promo_flag')['sales'].mean()
    plt.figure()
    plt.bar(['no promo','promo'], promo.values)
    plt.title('Promo uplift (mean sales)')
    plt.show()
```
